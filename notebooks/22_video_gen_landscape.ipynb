## Product Evolution Timeline

Chart the evolution of commercial video generation:

- **2023 Early**: Image-to-video (structure + style transfer). Key models: Gen-1, Stable Video Diffusion. Architecture: U-Net + temporal layers.
- **2023 Late**: Text-to-video (direct generation). Key models: Gen-2, Pika 1.0. Still short clips (4s), moderate quality.
- **2024 H1**: DiT-based scaling. Key moment: Sora demo (Feb 2024) — showed that scaling DiT + massive data = qualitative leap. Long-form (60s), high res, physical understanding.
- **2024 H2**: Production quality race. Gen-3 Alpha, Kling, Dream Machine. Focus on reliability, speed, controllability.
- **2025 H1**: Real-time and interactive. Autoregressive approaches enable streaming. CausVid (4-step causal generation). First interactive demos.
- **2025 H2**: World models. Moving beyond "generate a video" to "simulate a world." Physics-aware, interactive, multi-modal (video + audio + 3D). General World Models emerge.

## Architecture Evolution

Technical progression:

| Era | Architecture | Paradigm | Temporal | Limitation |
|-----|-------------|----------|----------|------------|
| 2022-23 | U-Net | DDPM/DDIM | Factored spatial+temporal attention | Slow, poor scaling |
| 2023-24 | DiT/MMDiT | Flow matching | 3D spacetime patches | Still slow, O(N²) attention |
| 2024-25 | DiT + distillation | Consistency/DMD/CausVid | Causal + bidirectional hybrid | Quality-speed tradeoff |
| 2025+ | AR + diffusion hybrid (A2D) | Autoregressive + diffusion refinement | Causal (streaming) | Error accumulation |

Key insight: the field is converging toward autoregressive generation with diffusion refinement (A2D). This enables streaming, interactivity, and scales with LLM infrastructure.

## Open Source Landscape

The open-source video generation ecosystem has exploded in 2024-2025. Key differentiators:

- **Wan2.1/2.2**: Largest open model, strong quality. Wan2.2 adds MoE for efficiency. Chinese-origin, Apache 2.0.
- **HunyuanVideo**: Strong data pipeline documentation (their paper is excellent on data curation). Dual-stream DiT.
- **Open-Sora 2.0**: Best documentation of data pipeline and training recipe. Good reference implementation.
- **LTX-Video 2**: Smallest model with competitive quality. Good for research on a budget.
- **CogVideoX**: 3D VAE + expert transformer. Good temporal consistency.

Key trend: open models are 6-12 months behind proprietary, but the gap is narrowing. Data quality is the primary differentiator.

## Data as Competitive Moat

Why data quality matters more than architecture at this point:

- All major players use similar architectures (DiT/MMDiT variants)
- Model scaling is well-understood (follow the LLM playbook)
- **Data quality is the primary differentiator**:
  - Caption quality: VLM re-captioning is now standard, but prompt engineering and quality filtering vary wildly
  - Data diversity: balanced representation across scenes, styles, motions, durations
  - Temporal quality: filtering for temporal consistency, smooth motion, no artifacts
  - Scale: proprietary datasets are 10-100x larger than public ones
  - The data flywheel: better data → better model → more users → more data (for companies with user-facing products)
- Dataset engineering is "transversal to everything" — it touches pre-training, finetuning, evaluation, and product quality

## World Models: The Next Frontier

The industry is converging on "world models" — AI systems that don't just generate video, but simulate worlds:

- **Core capability**: predict what happens next given current state + action
- **Applications**:
  - Creative tools: interactive scene exploration, real-time editing
  - Robotics: simulate robot actions before execution
  - Gaming: generate game worlds on-the-fly
  - Science: simulate physical processes
- **Technical requirements**:
  - Real-time generation (24+ FPS)
  - Interactive (condition on user actions)
  - Physics-aware (objects behave realistically)
  - Multi-modal (video + audio + haptics)
  - Long-context (maintain consistency over minutes)
- **Architecture implications**: autoregressive is natural (predict next state given current), but needs diffusion-level quality → A2D hybrid
- This is why AR video generation, step distillation, and long context are all critical research areas

## Key Questions to Consider

Thoughtful questions demonstrate engagement and technical depth. Consider these when evaluating teams, projects, or companies:

**Technical / Architecture:**
- "How are you thinking about the AR vs diffusion tradeoff for long-form generation?"
- "What's your approach to temporal consistency in videos longer than 10 seconds?"
- "How do you handle the quality-speed Pareto frontier for real-time applications?"

**Data / Infrastructure:**
- "What's your re-captioning strategy for video data? How do you ensure caption quality at scale?"
- "How does data quality feedback flow back from model evaluation to data curation?"
- "What's your approach to data mixing across different content domains?"

**Product / Strategy:**
- "How does the data team's work influence model architecture decisions?"
- "What does the feedback loop look like between research, data, and product?"
- "Where do you see the biggest leverage for improving model quality in the next 6 months?"

**Culture / Team:**
- "How do research and engineering collaborate on data quality?"
- "What does a typical project lifecycle look like from idea to production?"
- "What's the balance between exploration and exploitation in the research agenda?"

## Checkpoint

- **World models are the destination**: from "generate a video" to "simulate a world"
